## Visualizing World Data

#### Confirmed cases (Total Cases) all over the world

```python
df_world_over_time = df_world.groupby('date').sum()[['total_cases','new_cases','total_deaths','new_deaths']]
df_world_over_time
```

```python
df_world_over_time.reset_index(inplace=True)
```

```python
df_world_over_time
```

```python
df_world_over_time.info()
```

```python
df_world_over_time['date'] = pd.to_datetime(df_world_over_time['date'])
```

```python
df_world_over_time.info()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'])
plt.xlabel('Months')
plt.ylabel('Confirmed Total Cases')
plt.title('Confirmed Total Cases all over the world')
plt.xticks(rotation=90)
plt.show()
```

#### Total death cases evolution over time

```python
plt.figure(figsize=(20,5))
plt.plot(df_world_over_time['date'], df_world_over_time['total_deaths'], color = 'red')
plt.xlabel('Months')
plt.ylabel('Total Death cases')
plt.title('Total Death cases all over the world')
plt.xticks(rotation=90)
plt.show()
```

#### New Cases all over the world

```python
plt.figure(figsize=(20,5))
plt.plot(df_world_over_time['date'], df_world_over_time['new_cases'])
plt.xlabel('Months')
plt.ylabel('New Cases')
plt.title('New Cases all over the world')
plt.xticks(rotation=90)
plt.show()
```

#### Putting it all together

```python
plt.figure(figsize=(20,5))
#plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'])
plt.plot(df_world_over_time['date'], df_world_over_time['total_deaths'], label='Total Death Cases' ,color = 'red')
plt.plot(df_world_over_time['date'], df_world_over_time['new_cases'], label='New Cases')
plt.legend(loc = 0)
```
